```python
import pandas as pd
```

```python
GRD_2024 = pd.read_csv('GRD_PUBLICO_2024.csv', sep='|', encoding='latin-1', low_memory=False,parse_dates=['FECHA_NACIMIENTO', 'FECHA_INGRESO', 'FECHAALTA'], dayfirst=True, date_format='%d/%m/%Y')
GRD_2024['IR_29301_SEVERIDAD'] = pd.to_numeric(GRD_2024['IR_29301_SEVERIDAD'], errors='coerce')
GRD_2024['IR_29301_MORTALIDAD'] = pd.to_numeric(GRD_2024['IR_29301_MORTALIDAD'], errors='coerce')
GRD_2024['IR_29301_COD_GRD'] = pd.to_numeric(GRD_2024['IR_29301_COD_GRD'], errors='coerce')
GRD_2024['IR_29301_COD_GRD'] = GRD_2024['IR_29301_COD_GRD'].fillna(0)
GRD_2024['IR_29301_COD_GRD'] = GRD_2024['IR_29301_COD_GRD'].astype(str)
```

```python
establecimientos = pd.read_csv('establecimientos.csv', sep=';', low_memory=False)
establecimientosAlta = establecimientos[establecimientos['NivelComplejidadEstabGlosa'] == 'Alta Complejidad']
display(establecimientosAlta)
```

```python
GRD_2023 = pd.read_csv('GRD_PUBLICO_2023.txt', sep='|', encoding='utf-16', low_memory=False, decimal=',',parse_dates=['FECHA_NACIMIENTO', 'FECHA_INGRESO', 'FECHAALTA'], dayfirst=True, date_format='%d/%m/%Y')
GRD_2023['IR_29301_SEVERIDAD'] = pd.to_numeric(GRD_2023['IR_29301_SEVERIDAD'], errors='coerce')
GRD_2023['IR_29301_MORTALIDAD'] = pd.to_numeric(GRD_2023['IR_29301_MORTALIDAD'], errors='coerce')
GRD_2023['IR_29301_COD_GRD'] = pd.to_numeric(GRD_2023['IR_29301_COD_GRD'], errors='coerce')
GRD_2023['IR_29301_COD_GRD'] = GRD_2023['IR_29301_COD_GRD'].fillna(0)
GRD_2023['IR_29301_COD_GRD'] = GRD_2023['IR_29301_COD_GRD'].astype(str)

```

```python
# leyendo diccionarios
hospitales = pd.read_csv('Hospitales.csv', sep='|', encoding='utf-8', header=None, names=['COD_HOSPITAL', 'NOMBRE_HOSPITAL'])
irgrd = pd.read_csv('IR-GRD.csv', sep='|',  header=None, names=['COD_GRD', 'DESCRIPCION_GRD'])
mortalidad = pd.read_csv('Mortalidad.csv', sep='|', encoding='utf-8',  header=None, names=['CODIGO', 'SEVERIDAD'])
severidad = pd.read_csv('Severidad.csv', sep='|', encoding='utf-8', header=None, names=['CODIGO', 'MORTALIDAD'])
irgrd['COD_GRD'] = irgrd['COD_GRD'].astype(str)
irgrd['GRD_CODIGO_LIMPIO'] = irgrd['COD_GRD'].str.extract(r'(\d+)')


GRD_2023_new = GRD_2023.merge(hospitales, on='COD_HOSPITAL', how='left')

GRD_2023_new['GRD_CODIGO_LIMPIO'] = GRD_2023_new['IR_29301_COD_GRD'].str.extract(r'(\d+)')

GRD_2023_new = GRD_2023_new.merge(irgrd[['GRD_CODIGO_LIMPIO', 'DESCRIPCION_GRD']], on='GRD_CODIGO_LIMPIO', how='left')

# Severidad y mortalidad
GRD_2023_new = GRD_2023_new.merge(severidad, left_on='IR_29301_SEVERIDAD', right_on='CODIGO', how='left')
GRD_2023_new = GRD_2023_new.merge(mortalidad, left_on='IR_29301_MORTALIDAD', right_on='CODIGO', how='left', suffixes=('_sev', '_mor'))




# Repetimos el proceso para GRD_2024

GRD_2024_new = GRD_2024.merge(hospitales, on='COD_HOSPITAL', how='left')

GRD_2024_new['GRD_CODIGO_LIMPIO'] = GRD_2024_new['IR_29301_COD_GRD'].str.extract(r'(\d+)')

GRD_2024_new = GRD_2024_new.merge(irgrd[['GRD_CODIGO_LIMPIO', 'DESCRIPCION_GRD']], on='GRD_CODIGO_LIMPIO', how='left')

# AÃ±adimos descripciones de severidad y mortalidad
GRD_2024_new = GRD_2024_new.merge(severidad, left_on='IR_29301_SEVERIDAD', right_on='CODIGO', how='left')
GRD_2024_new = GRD_2024_new.merge(mortalidad, left_on='IR_29301_MORTALIDAD', right_on='CODIGO', how='left', suffixes=('_sev', '_mor'))
```

```python
#Buscando si el id de hospital de GRD_2024_new son los mismos que los de establecimientos
GRD_2024_new['COD_HOSPITAL'].isin(establecimientosAlta['EstablecimientoCodigo']).value_counts()
# Mostramos codigos en ambos datasets
codigos_comunes = GRD_2024_new['COD_HOSPITAL'][GRD_2024_new['COD_HOSPITAL'].isin(establecimientosAlta['EstablecimientoCodigo'])].unique()
codigos_comunes

```

```python
GRD_2024_ALTA = GRD_2024_new[GRD_2024_new['COD_HOSPITAL'].isin(establecimientosAlta['EstablecimientoCodigo'])]
GRD_2023_ALTA = GRD_2023_new[GRD_2023_new['COD_HOSPITAL'].isin(establecimientosAlta['EstablecimientoCodigo'])]

# GRD con ambos aÃ±os y solo hospitales de alta complejidad con su latitud y longitud
GRD_ALTA = pd.concat([GRD_2023_ALTA, GRD_2024_ALTA], ignore_index=True)
GRD_ALTA = GRD_ALTA.merge(establecimientosAlta[['EstablecimientoCodigo', 'Latitud', 'Longitud']], left_on='COD_HOSPITAL', right_on='EstablecimientoCodigo', how='left')
GRD_ALTA.drop(columns=['EstablecimientoCodigo'], inplace=True)

```

```python
import matplotlib.pyplot as plt

# top 10 provincias con mÃ¡s casos de hospitales de alta complejidad
top_regiones = GRD_ALTA['PROVINCIA'].value_counts().head(10)
plt.figure(figsize=(10,6))
top_regiones.plot(kind='bar')
plt.title('NÃºmero de Casos por Provincia en Hospitales de Alta Complejidad')
plt.xlabel('Provincia')
plt.ylabel('NÃºmero de Casos')
plt.xticks(rotation=45)
plt.show()
```
